```python
#Gene expression data
genes_data = pd.read_csv('heatmapExpressionData.csv')
display(genes_data)
```

```python
genes = genes_data
```

```python
genes.set_index('GeneName', inplace=True)

plt.figure(figsize=(12, 8)) #figsize 25,30 adequate to display all genes here
sns.heatmap(genes, cmap='viridis',linewidths=0.5) #,yticklabels=True 
plt.title('Heatmap of Gene Expression Data')
plt.xlabel('Samples')
plt.ylabel('Genes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DG4', y='DG1', data=genes)
plt.title('Scatter Plot of Gene Expression')
plt.grid(True)

# Show the plot
plt.show()
```

```python
transposed_genes = genes.set_index('GeneName').transpose()

# Plot a line plot of the gene expression data
plt.figure(figsize=(12, 8))
#for gene in transposed_genes.columns:
    #plt.plot(transposed_genes.index, transposed_genes[gene], label=gene)
transposed_genes.plot(legend=False, alpha=0.5)
plt.title('Line Plot of Gene Expression Data')
plt.xlabel('Samples')
plt.ylabel('Gene Expression')
#plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
#plt.grid(True)
plt.show()
```

```python
#to see the genes showing different expression in one sample 

sample_name = 'NAc2'
threshold = 3.0  # Expression level threshold

# Identify genes with expression levels above the threshold for the specified sample
genes_above_threshold = genes.loc[genes[sample_name] > threshold]

# Reset the index to make 'GeneName' a column again
#genes_above_threshold.reset_index(inplace=True)

# Print the genes with expression levels above the threshold for the specified sample
print("Genes with expression levels above {} for sample '{}':".format(threshold, sample_name))
display(genes_above_threshold)
#display(genes.tail(55))
```
